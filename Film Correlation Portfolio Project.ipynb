```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Siddharth\Desktop\movies.csv')
```

```python
df.head()
```

```python
##Let's see if there is any missing data
for col in df.columns:
    ValuePercent_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, ValuePercent_missing))
```

```python
data= df.dropna()
```

```python
data
```

```python
##Data types for our column
data.dtypes
```

```python
##Converting the datatype for column gross and budget from float to int.
data = data.astype({'budget' : int, 'gross' : int})
data
```

```python
##Creating new column yearcorrect and putting the value (year) that is extracted from the released column.
data['yearcorrect'] = data['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
data
```

```python
##To display all the rows

##pd.set_option('display.max_rows', None)
```

```python
##Drop any duplicates

data['name'].drop_duplicates().sort_values()
```

```python
##Deriving the correlation.
##Scatter plot with budget and gross

plt.scatter(x=data['budget'], y=data['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget of Film')
plt.ylabel('Gross Earning from Film')
plt.show()
```

```python
data=data.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
data.head()
```

```python
print(data.loc[data['name'] == 'Avatar'])
```

```python
##Plotting budget vs gross using seasborn

sns.regplot(x='budget', y='gross', data=data, scatter_kws={"color" : "red"}, line_kws={"color" : "blue"})
```

```python
##Let's start looking at correlation

data.corr(method="pearson") #pearson,kindall,spearman
```

```python
correlation_matrix = data.corr(method="pearson")

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Feature")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
data_numerized = data.copy()

for col_name in data_numerized.columns:
    if(data_numerized[col_name].dtype == 'object'):
        data_numerized[col_name] = data_numerized[col_name].astype('category')
        data_numerized[col_name]= data_numerized[col_name].cat.codes
        
data_numerized
```

```python
correlation_matrix = data_numerized.corr(method="pearson")

sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Matrix for Numeric Feature")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
data_numerized.corr()
```

```python
##Using Unstack

correlation_mat = data_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
